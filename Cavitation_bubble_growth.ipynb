# Data of calculated pressure (Fig. 2.5)
File: 'cav_p_calc_Gly80.xls'

```python
df_p = pd.read_excel('cav_p_calc_Gly80.xls')
df_p
```

```python
plt.plot(df_p.t_no_cavitation,df_p.p_no_cavitation,'bo')
plt.plot(df_p.t_no_cavitation_1,df_p.p_no_cavitation_1,'bo')

plt.plot(df_p.t_cavitation,df_p.p_cavitation,'go')
plt.plot(df_p.t_cavitation_1,df_p.p_cavitation_1,'go')

plt.plot(df_p.t_transient_cavitation,df_p.p_transient_cavitation,'ro')
plt.plot(df_p.t_transient_cavitation_1,df_p.p_transient_cavitation_1,'ro')
plt.plot(df_p.t_transient_cavitation_2,df_p.p_transient_cavitation_2,'ro')

plt.yscale('log')
plt.xlim([0,0.0015]);
plt.ylim([1000,100000000]);
```
